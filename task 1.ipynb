```python
import pandas as pd 
import matplotlib.pyplot as plt
# Load the data from the CSV file
file_path = r'C:\Users\VIDHUN\Downloads\csv file\API_SP.POP.TOTL_DS2_en_csv_v2_6560098 (1).csv'
data = pd.read_csv(file_path, skiprows=4)
data
```

```python
pr= data[['Country Name','Country Code','Indicator Name','1960','1961','1962','1963','1964','1965','1966','1967','1968','1969','2013','2014','2015','2016','2017','2018','2019','2020','2021']]
pr.head(10)
```

```python

# Extract the relevant columns for analysis (e.g., Country and Total Population for different years)
# Replace 'CountryColumn' with the actual column name for countries
# You can check the column names using data.columns
country_data = pr['Country Name']
population_data = pr.drop(['Country Name', 'Country Code', 'Indicator Name'], axis=1)

# Transpose the data for better visualization (countries in rows, years in columns)
population_data = population_data.T

# Create a horizontal bar chart based on the data
plt.figure(figsize=(15, 8))  # Adjust the figure size as needed

# Plot horizontal bar chart without legend
ax = population_data.plot(kind='barh', stacked=True, colormap='viridis', width=0.8, legend=False)

# Customize the plot as needed
plt.xlabel('Total Population')
plt.ylabel('Year')
plt.title('Total Population of Countries Over the Years')

# Improve spacing between years on the x-axis
years = population_data.index
ax.set_yticks(range(len(years)))
ax.set_yticklabels(years)

# Show the plot
plt.show()
```
